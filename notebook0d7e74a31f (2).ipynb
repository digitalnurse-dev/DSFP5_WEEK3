```python
#create a dictionary 'data' with 3 key value pairs - name , age and score
Data = {
    "name": ["sandy", "nara", "hanna", "mike", "Amira"],
    "age": ["19", "20", "16", "15", "17"],
    "score": ["70", "55", "80", "45","92"]
}
print(Data)
```

```python
#using the dictionary create a dataframe
Data = {
    "name": ["sandy", "nara", "hanna", "mike", "Amira"],
    "age": [19, 20, 16, 15, 17],
    "score": [70, 55, 80, 45, 92]
}
df = pd.DataFrame(Data)
print(df)
```

```python
#filter int
df['age'] > 18
```

```python
#to filter only true values
df[df['age'] > 18]
```

```python
#filter strings
df[df['name'] == 'mike']
```

```python
#to filter more than one element
df[df['name'].isin(['mike','hanna'])]
```
